# creating dummies for categorical value

```python
Gender1=pd.get_dummies(df["Gender"],drop_first =True)
Gender1.head()
```

```python
Transport1=pd.get_dummies(df["Transport"],drop_first =True)
Transport1.head()
```

```python
df =pd.concat([df,Gender1,Transport1],axis=1)
df.head()
```

```python
df.drop(["Gender","Transport"],axis=1,inplace=True)
df.head()
```

# Split the Data & fit the model

```python
df.head()
```

```python
X = df.drop('Public Transport',axis=1)

Y = df['Public Transport']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.30, random_state=123)
```

```python
X_train.head()
```

```python
# Scaling the X_train & X_test

from sklearn.preprocessing import StandardScaler

ss = StandardScaler() 

X_train_scaled = pd.DataFrame(ss.fit_transform(X_train) , columns = X_train.columns) 


X_test_scaled = pd.DataFrame(ss.transform(X_test) , columns = X_test.columns) 
```

```python
# Model Train

from sklearn.neural_network import MLPClassifier

nncl = MLPClassifier(max_iter=2000, 
                     verbose = True , 
                     hidden_layer_sizes= (100,100,100), 
                     activation='logistic' , 
                     solver = 'sgd' , 
                     batch_size=100, 
                     learning_rate='invscaling',
                     learning_rate_init = 0.001,
                     tol=.00001) 
```

```python
nncl.fit(X_train_scaled,Y_train)
```

```python
nncl.score(X_train_scaled,Y_train)
```

```python
nncl.score(X_test_scaled,Y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred_train = nncl.predict(X_train_scaled)
y_pred_test = nncl.predict(X_test_scaled)

print(classification_report(Y_train,y_pred_train))
print(classification_report(Y_test,y_pred_test))
```

```python
nncl = MLPClassifier(max_iter=2000, 
                     verbose = True , 
                     hidden_layer_sizes= (100,), 
                     activation='relu' , 
                     solver = 'adam' , 
                     batch_size='auto', 
                     learning_rate='invscaling',
                     learning_rate_init = 0.001,
                     tol=.00001) 
```

```python
nncl.fit(X_train_scaled,Y_train)
```

```python
print(nncl.score(X_train_scaled,Y_train))
print(nncl.score(X_test_scaled,Y_test))
```
